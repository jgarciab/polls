```python
import pandas as pd
import numpy as np
import seaborn as sns
import pylab as plt

%matplotlib inline
```

```python
spain = pd.read_csv("data/all_tables_Spain_cleaned.csv",index_col=0,parse_dates=["date"])
spain = spain.sort_values(by="date")
#spain = spain.dropna(thresh=len(spain)/6,axis=1)
spain.head()
```

```python
import numpy as np
import statsmodels.api as sm
lowess = sm.nonparametric.lowess


plt.figure(figsize=(20,4))
plt.axis("on")

color = ["b","r","purple","orange","green"]
i = 0
for party in ["PP","PSOE","Podemos","Ciudadanos","IU"]:
    pp = lowess(spain[party],spain["date"],return_sorted=False,frac=0.1)
    plt.plot(spain["date"],spain[party],",",color=color[i])
    plt.plot(spain["date"],pp,color=color[i],label=party)
    i += 1
plt.legend()
#

```
